```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZF1DX0XUsuxWHRQd"
playlist_URI = playlist_link.split("/")[-1].split("?")[0]
track_uris = [x["track"]["id"] for x in sp.playlist_tracks(playlist_URI)["items"]]

ids= []

for track in sp.playlist_tracks(playlist_URI)["items"]:
    #URI
    ids.append(track["track"]["id"])
```

```python

def getTrackFeatures(id):
  track_info = sp.track(id)
  features_info = sp.audio_features(id)

  # Track info
  name = track_info['name']
  id = track_info["id"]
  album = track_info['album']['name']
  album_id = track_info['album']['id']
  artist = track_info['album']['artists'][0]['name']
  release_date = track_info['album']['release_date']
  length = track_info['duration_ms']
  popularity = track_info['popularity']
  genre = 'country'
  # Track features
  acousticness = features_info[0]['acousticness']
  danceability = features_info[0]['danceability']
  energy = features_info[0]['energy']
  instrumentalness = features_info[0]['instrumentalness']
  liveness = features_info[0]['liveness']
  loudness = features_info[0]['loudness']
  speechiness = features_info[0]['speechiness']
  tempo = features_info[0]['tempo']
  time_signature = features_info[0]['time_signature']
  key = features_info[0]['key']
  valence = features_info[0]['valence']
  track_data = [name, id, album, album_id, artist, release_date, length, popularity,acousticness, danceability, energy, instrumentalness, liveness, loudness, speechiness, tempo, time_signature, key, valence, genre]
  return track_data
```

# New Section

```python
track_list = []
for i in range(len(ids)):
  full_data = getTrackFeatures(ids[i])
  track_list.append(full_data)
full_df = pd.DataFrame(track_list, columns = ['Name', "track_id", 'Album',"albumid",  'Artist', 'Release_date', 'Length', 'Popularity', 'Acousticness', 'Danceability', 'Energy', 'Instrumentalness', 'Liveness', 'Loudness', 'Speechiness', 'Tempo', 'Time_signature', "key", "valence", "genre"])
full_df.head()
```
